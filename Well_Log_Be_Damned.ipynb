```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
shrimplin = pd.read_csv('shrimplin.csv')
```

```python
shrimplin.describe()
```

```python
#'GR', 'Depth', data = shrimplin
#shrimplin['GR'], shrimplin['Depth'], color = 'red'
```

```python
f, (ax1, ax2, ax3, ax4, ax5, ax6, ax7) = plt.subplots(1, 7, sharey=True, figsize = (20,10))
f.suptitle('Bitch Ass Well Plots')

ax1.plot('GR', 'Depth', data = shrimplin, color = 'red')
ax1.set(xlabel = 'GR')
ax1.grid()
ax2.plot(shrimplin['ILD_log10'], shrimplin['Depth'], color = 'orange')
ax2.set(xlabel = 'ILD_log10')
ax2.grid()
ax3.plot(shrimplin['DeltaPHI'], shrimplin['Depth'], color = 'yellow')
ax3.set(xlabel = 'DeltaPHI')
ax3.grid()
ax4.plot(shrimplin['PHIND'], shrimplin['Depth'], color = 'green')
ax4.set(xlabel = 'PHIND')
ax4.grid()
ax5.plot(shrimplin['PE'], shrimplin['Depth'], color = 'blue')
ax5.set(xlabel = 'PE')
ax5.grid()
ax6.plot(shrimplin['NM_M'], shrimplin['Depth'], color = 'purple')
ax6.set(xlabel = 'NM_M')
ax6.grid()
ax7.plot(shrimplin['RELPOS'], shrimplin['Depth'], color = 'black')
ax7.set(xlabel = 'RELPOS')
ax7.grid()
```
